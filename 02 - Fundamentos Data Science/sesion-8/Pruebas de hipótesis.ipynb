## 3. Implemente una función de prueba de hipótesis
La función debe ingresar los siguientes argumentos:
* df : La tabla de datos.
* variable : La variable a analizar.
* binarize : El indicador binario a utilizar.

tips:
* Separe la variable en dos, utilizando el indicador binario. Recuerde eliminar los perdidos con dropna() .
*  Implemente ttest_ind de scipy.stats y guarde el valor t y pval .
*  Reporte las medias para cada grupo (0 y 1).
* Reporte la diferencia de entre las medias.
* Reporte los valores t y p

```python
from scipy import stats

def hip_test(dataframe, variable, binarize):
    group_1 = dataframe[dataframe[binarize] == 1][variable].dropna()
    group_0 = dataframe[dataframe[binarize] == 0][variable].dropna()

    t, pval = stats.ttest_ind(group_1, group_0)

    mean_report = 'Media grupo {}: {}'
    mean_0 = group_0.mean()
    mean_1 = group_1.mean()

    print(mean_report.format(0, mean_0))
    print(mean_report.format(1, mean_1))
    diff = abs(mean_1 - mean_0)
    r = max(group_0.max(), group_1.max()) - min(group_0.min(), group_1.min())
    percent_diff = round(100 * diff / r)
    print('Diferencia medias: {} ({}%)'.format(diff, percent_diff))

    print('t: {}, p: {}'.format(t, pval))
    
    if abs(t) < 1.96:
        print('No hay suficiente evidencia estadística para inferir que la hipótesis nula es falsa')
    else:
        print('Hay suficiente evidencia estadística para inferir que la hipótesis nula es falsa')
        
    if pval > 0.05:
        print('p > 0.05, fallamos en rechazar la hipótesis nula')
    else:
        print('p < 0.05, rechazamos la hipótesis nula')
```

```python
# Vamos a reportar para los atributos correspondientes a mi apellido

for region in df['region'].unique():
    for variable in ['chldmort', 'adfert', 'life']:
        binarize = region.lower()
        print('== Region: {}, Atributo: {} =='.format(region, variable))
        hip_test(df, variable, binarize)
        print('\n')
```

## 4. Implemente una función que grafique los histogramas para ambas muestras.

Genere una función que devuelva un gráfico donde visualice las dos histogramas cuando la
variable es 1 y 0, respectivamente.

tips: 
* Incluya el método hist de matplotlib.pyplot.
* Incluya los argumentos alpha y label .
* Para las tres variables de interés acorde a su grupo, analice las diferencias de medias por cada continente, y posteriormente grafique. Concluya con los principales resultados al respecto.

```python
import matplotlib.pyplot as plt
```

```python
def graph_hist(dataframe, variable, binarize):
    hist_1 = dataframe[dataframe[binarize] == 1][variable].dropna()
    hist_0 = dataframe[dataframe[binarize] == 0][variable].dropna()
    
    def draw(hist, hist_name, color):
        plt.hist(hist, alpha=0.5, color=color, density=True, label='{} {}'.format(binarize, hist_name))
        plt.axvline(np.mean(hist), color=color)

    draw(hist_1, '1', 'tomato')
    draw(hist_0, '0', 'lightblue')
    plt.title('{} {}'.format(binarize, variable))
    plt.legend(loc='best')
    plt.show()
```

```python
for region in df['region'].unique():
    for variable in ['chldmort', 'adfert', 'life']:
        binarize = region.lower()
        print('== Region: {}, Atributo: {} =='.format(region, variable))
        hip_test(df, variable, binarize)
        graph_hist(df, variable, binarize)
        print('\n')
```
